```python
# climate is missing for few countries but not the region , lets use this to our advantage by fill in the missing Climate values based on the mean climate value for its region
```

```python
df['Climate'] = df['Climate'].fillna(df.groupby('Region')['Climate'].transform('mean'))
```

```python
# literacy (%)
df['Literacy (%)'] = df['Literacy (%)'].fillna(df.groupby('Region')['Literacy (%)'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.info()
```

## Data Preparation

```python
X = df.drop('Country',axis=1)
```

```python
X = pd.get_dummies(X)
```

```python
X
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X
```

```python
from sklearn.cluster import KMeans
```

```python
ssd = []

for k in range(2,30):
    model =KMeans(n_clusters = k)
    model.fit(scaled_X)

    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,30),ssd,'o--')
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```

```python
model = KMeans(n_clusters = 3)
model.fit(scaled_X)
```

```python
X['K=3 Cluster labels'] = model.labels_
```

```python
X.corr()['K=3 Cluster labels'].sort_values().plot(kind='bar')
```
